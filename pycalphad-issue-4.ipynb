```python
%matplotlib inline
from matplotlib import pyplot as plt
from pycalphad import Database, ternplot, variables as v
from time import time
```

```python
dbfile = '/Users/brandon/Box Sync/databases/nims-databases/Cr-Ti-V.tdb'
dbf = Database(dbfile)
comps = ['CR', 'TI', 'V', 'VA']
phases = list(dbf.phases.keys())
conditions = {v.X('TI'): (0,1,0.01), v.X('V'): (0,1,0.01), v.P: 101325, v.T: 873}
```

```python
start = time()
fig = plt.figure(figsize=(10,10))
ternplot(dbf, comps, phases, conditions, x=v.X('V'), y=v.X('TI'))
print('Walltime: {}'.format(time()-start))
```
